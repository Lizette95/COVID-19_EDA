### Import and clean 'local cases' dataset

```python
# Import 'local cases' dataset
url = "https://9gnht4xyvf.execute-api.eu-west-1.amazonaws.com/api/get_table/CasesLocalView" 
headers = {'x-api-key':"XXX"}
response = requests.request("GET", url, headers=headers)
x = response.json()
local_cases = pd.DataFrame(x)

# Display
local_cases.head()
```

```python
# Drop unnecessary columns
local_cases = local_cases.drop(['id','country_id','location_level'],axis=1)

# Change data types where necessary
local_cases.dtypes
local_cases['date'] = pd.to_datetime(local_cases['date'])

# Replace missing (zero) values for plotting
# See empty cells in https://github.com/dsfsi/covid19za/blob/master/data/covid19za_provincial_cumulative_timeline_confirmed.csv
local_cases.loc[(local_cases.date == '2020-03-27'),'confirmed'] = np.nan
local_cases.loc[(local_cases.date == '2020-04-07'),'confirmed'] = np.nan

# Keep cases where the province is known
known = local_cases[(local_cases['location']!='UNKNOWN') & (local_cases['location']!='source')]
provinces = ['Eastern Cape','Free State','Gauteng','KwaZulu-Natal','Limpopo','Mpumalanga','Northern Cape','North West','Western Cape']
province_palette = {'EC':'#F1EC52','FS':'#B1DC36','GP':'#FCAB64','KZN':'#61CD2E','LP':'#8E559B','MP':'#6DAEE2','NC':'#4AAD95','NW':'#5DD39E','WC':'#EA96A3'}

# Group by province
by_province = known.groupby('location').max()
total_per_province = by_province['confirmed'].sort_values(ascending=False).reset_index()
total_per_province['confirmed'] = total_per_province['confirmed'].astype(int)

# Display
total_per_province
```

<a id="growth_l"></a>
### Increase in local COVID-19 cases over time

```python
# Line plot of increase in confirmed cases over time per province
fig,ax = plt.subplots(figsize=(10,5))
sns.lineplot(x='date',y='confirmed',hue='location',data=known,palette=province_palette)
ax.xaxis.set_major_formatter(DateFormatter("%d %b"))
plt.title('Increase in Local COVID-19 Cases Over Time',fontsize=16)
plt.xlabel('Date',fontsize=12)
plt.ylabel('Number of Confirmed COVID-19 Cases',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(provinces,title='Province',frameon=False,fontsize=12,title_fontsize=12)

# Display
plt.show()
```

<a id="total_l"></a>
### Total confirmed COVID-19 cases per province

```python
# Bar plot of confirmed cases per province
fig,ax = plt.subplots(figsize=(10,5))
bar1 = sns.barplot(x='location',y='confirmed',data=total_per_province,palette=province_palette)
plt.title('Confirmed COVID-19 Cases Per Province',fontsize=16)
plt.xlabel('Province',fontsize=12)
plt.ylabel('Number of Confirmed COVID-19 Cases',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for index, row in total_per_province.iterrows():
    bar1.text(row.name,row.confirmed,row.confirmed,ha="center",fontsize=12)

# Display
plt.show()
```

When we look at the total number of COVID-19 cases in South Africa, we see that Gauteng has the highest number of cases.  
However, to determine the severity of the outbreak per province, we need to account for population size.

<a id="permil_l"></a>
### Confirmed COVID-19 cases per million people per province

```python
# Load population data per province & merge with 'local cases' dataset
# From https://www.statssa.gov.za/publications/P0302/P03022019.pdf
data = {'location':['GP','KZN','WC','EC','LP','MP','NW','FS','NC'],
        'population':[15176115,11289086,6844272,6712276,5982584,4592187,4027160,2887465,1263875]}
population = pd.DataFrame(data,columns = ['location', 'population'])
merged = pd.merge(population,total_per_province,on='location')

# Add column for cases per million people
merged['cases_per_million'] = (merged['confirmed'] // (merged['population']/1000000)).astype(int)
merged = merged.sort_values(by='cases_per_million',ascending=False).reset_index(drop=True)

# Display
merged
```

```python
# Bar plot of confirmed cases per million people in each province
fig,ax = plt.subplots(figsize=(10,5))
bar2 = sns.barplot(x='location',y='cases_per_million',data=merged,palette=province_palette)
plt.title('Confirmed COVID-19 Cases Per Million People Per Province',fontsize=16)
plt.xlabel('Province',fontsize=12)
plt.ylabel('Confirmed COVID-19 Cases Per Million People',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for index, row in merged.iterrows():
    bar2.text(row.name,row.cases_per_million,row.cases_per_million,ha="center",fontsize=12)

# Display
plt.show()
```

When population size is taken into account, we see that the Western Cape has a much higher number of COVID-19 cases per million people than Gauteng.

<a id="tests"></a>
### Import and clean 'testing' dataset

```python
# Import 'testing' dataset
url = "https://9gnht4xyvf.execute-api.eu-west-1.amazonaws.com/api/get_table/Testing" 
headers = {'x-api-key':"XXX"}
response = requests.request("GET", url, headers=headers)
x = response.json()
tests = pd.DataFrame(x)

# Display
tests.head()
```

```python
# Drop unnecessary columns
tests = tests.drop('country',axis=1)
cases = local_cases.drop(['location'],axis=1)

# Merge 'cases' and 'tests' dataset
tests = tests.set_index('date')
cases = cases.groupby('date').sum()
cases = cases.reset_index().set_index('date')
cases_vs_tests=pd.merge(cases,tests, how='inner', left_index=True, right_index=True)
cases_vs_tests = cases_vs_tests.reset_index()

# Replace missing (zero) values for plotting
# See empty cells in https://github.com/dsfsi/covid19za/blob/master/data/covid19za_provincial_cumulative_timeline_confirmed.csv
# See empty cells in https://github.com/dsfsi/covid19za/blob/master/data/covid19za_timeline_testing.csv
cases_vs_tests.loc[(cases_vs_tests.date == '2020-03-25'),'cumulative_tests'] = np.nan
cases_vs_tests.loc[(cases_vs_tests.date == '2020-03-27'),['cumulative_tests','confirmed']] = np.nan
cases_vs_tests.loc[(cases_vs_tests.date == '2020-04-07'),'confirmed'] = np.nan

# Display
cases_vs_tests.head()
```

<a id="cases_vs_tests"></a>
### Confirmed COVID-19 cases vs. number of tests administered

```python
# Joint plot of confirmed cases vs. tests administered
joint1 = sns.jointplot(x='cumulative_tests', y='confirmed',data=cases_vs_tests, kind='scatter',color='#0F75BC')
joint1.set_axis_labels('Number of Tests Administered', 'Number of Confirmed COVID-19 Cases', fontsize=12)
plt.subplots_adjust(top=0.9)
plt.suptitle('Confirmed COVID-19 Cases vs. Tests Administered',fontsize=16)

# Calculate Pearson correlation coefficient
joint1.annotate(stats.pearsonr,frameon=False,fontsize=12)

# Display
plt.show()
```

As is expected, we see that the number of confirmed cases increases as the number of tests administered increases.  
Due to this, is important to consider that we will see a significant increase in the number of confirmed cases once rate of testing in South Africa increases.

### Import and clean 'global cases' dataset

```python
# Import 'global cases' dataset
url = "https://9gnht4xyvf.execute-api.eu-west-1.amazonaws.com/api/get_table/CasesGlobalView" 
headers = {'x-api-key':"XXX"}
response = requests.request("GET", url, headers=headers)
x = response.json()
global_cases = pd.DataFrame(x)

# Change data types where necessary
global_cases.dtypes
global_cases['date'] = pd.to_datetime(global_cases['date'])

# Display
global_cases.head()
```

<a id="total_g"></a>
### Total number of COVID-19 cases worldwide

```python
# Group by country
keep = ['country','lat','long','confirmed','deaths','recovered','active']
by_country = global_cases[keep].groupby('country').max().reset_index()

# Check statistics
print(by_country.iloc[:,3:].describe().astype(int))

# Display
by_country.head()
```

```python
# Calculate case totals
totals = pd.DataFrame(by_country[['confirmed','deaths','recovered','active']].sum()).reset_index()
totals = totals.rename(columns={'index':'cases',0:'total'}).sort_values(by='total',ascending=False).reset_index()

# Bar plot of worldwide cases
fig,ax = plt.subplots(figsize=(6,5))
bar3 = sns.barplot(x='cases',y='total',data=totals,palette='bright')
plt.title('Total Number of COVID-19 Cases Worldwide',fontsize=16)
plt.xlabel('Type',fontsize=12)
plt.ylabel('Number of cases worldwide',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for index, row in totals.iterrows():
    bar3.text(row.name,row.total,row.total,ha="center",fontsize=12)

# Display
plt.show()
```

<a id="growth_g"></a>
### Increase in global COVID-19 cases over time

```python
# Group by date
keep = ['confirmed','active','recovered','deaths']
by_date = global_cases.groupby(['date','country'])[keep].max().reset_index()
by_date = by_date.groupby(['date'])[keep].sum().reset_index()

# Check statistics
print(by_date.describe().astype(int))

# Display
by_date.head()
```

```python
# Line plot of increase in global cases over time
fig,ax = plt.subplots(figsize=(12,6))
line = sns.lineplot(x='date',y='value', hue='variable',data=pd.melt(by_date, ['date']),palette='bright')
ax.xaxis.set_major_formatter(DateFormatter("%d %b"))
plt.title('Increase in Global COVID-19 Cases Over Time',fontsize=16)
plt.xlabel('Date',fontsize=12)
plt.ylabel('Number of COVID-19 Cases',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Type', labels=['Confirmed','Active','Recovered','Deaths'],frameon=False,fontsize=12,title_fontsize=12)

# Display
plt.show()
```

<a id="global_pop"></a>
### Import, clean & merge 'global population' dataset

```python
# Import 'global population' dataset
url = 'https://www.worldometers.info/geography/alphabetical-list-of-countries/'
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page, "html.parser")
my_table = soup.find('table',{'class':'table table-hover table-condensed'})
headers = [item.text for item in soup.find_all('th')]
table_rows = my_table.find_all('tr')
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [tr.text for tr in td]
    l.append(row)
global_population = pd.DataFrame(l, columns=headers)

# Display
global_population.head()
```

```python
# Drop unnecessary columns/rows
global_pop = global_population.drop(['#'],axis=1)
global_pop = global_pop.drop(global_pop.index[0])

# Fix country names
global_pop.iloc[24, 0] = 'Brunei'
global_pop.iloc[39, 0] = 'Congo (Brazzaville)'
global_pop.iloc[44, 0] = 'Czechia'
global_pop.iloc[45, 0] = 'Congo (Kinshasa)'
global_pop.iloc[56, 0] = 'Eswatini'
global_pop.iloc[118, 0] = 'Burma'
global_pop.iloc[186, 0] = 'US'

# Rename columns
global_pop = global_pop.rename(columns={'Country':'country','Population(2020)':'population','Land Area (Km²)':'area','Density(P/Km²)':'density'})

# Change data types where necessary
global_pop.dtypes
global_pop[['population','area','density']] = global_pop[['population','area','density']].replace(',', '', regex=True)
global_pop[['population','area','density']] = global_pop[['population','area','density']].apply(pd.to_numeric)

# Display
print(global_pop.shape)
global_pop.head()
```

```python
# Join 'global population' & 'global cases' dataset
joined = pd.merge(global_pop,by_country,how='left',on='country')

# Check statistics
print(joined.describe().astype(int))
joined.fillna(0, inplace=True)
joined.isnull().sum()

# Add columns for cases and deaths per million people per country
joined['cases_per_mil'] = joined['confirmed'] // (joined['population']/1000000)
joined['deaths_per_mil'] = joined['deaths'] // (joined['population']/1000000)
joined.iloc[:,6:] = joined.iloc[:,6:].astype(int)

# Display
joined.head()
```

<a id="top10_cases"></a>
### Top 10 countries with the highest number of COVID-19 cases

```python
# Bar plot of worldwide cases
top10_confirmed = joined.sort_values(by='confirmed',ascending=False).reset_index().head(10)
fig,ax = plt.subplots(figsize=(15,6))
bar4 = sns.barplot(x='country',y='confirmed',data=top10_confirmed,palette='husl')
plt.title('Top 10 Countries With the Highest Number of COVID-19 Cases',fontsize=16)
plt.xlabel('Country',fontsize=12)
plt.ylabel('Number of COVID-19 Cases',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for index, row in top10_confirmed.iterrows():
    bar4.text(row.name,row.confirmed,row.confirmed,ha="center",fontsize=12)

# Display
plt.show()
```

When we look at the total number of global COVID-19 cases, we see that the US has the highest number of cases.  
However, as with the local cases per province, we also need to account for the population size per country.

<a id="top10_casespermil"></a>
### Top 10 countries with the highest number of COVID-19 cases per million people

```python
# Bar plot of worldwide cases
top10_cases_per_mil = joined[joined['population']>1000000].sort_values(by='cases_per_mil',ascending=False).reset_index().head(10)
fig,ax = plt.subplots(figsize=(15,6))
bar5 = sns.barplot(x='country',y='cases_per_mil',data=top10_cases_per_mil,palette='husl')
plt.title('Top 10 Countries With the Highest Number of COVID-19 Cases Per Million People',fontsize=16)
plt.xlabel('Country',fontsize=12)
plt.ylabel('Number of COVID-19 Cases Per Million People',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for index, row in top10_cases_per_mil.iterrows():
    bar5.text(row.name,row.cases_per_mil,row.cases_per_mil,ha="center",fontsize=12)

# Display
plt.show()
```

### Top 10 countries with the highest number of COVID-19 deaths

```python
# Bar plot of worldwide deaths
top10_deaths = joined.sort_values(by='deaths',ascending=False).reset_index().head(10)
fig,ax = plt.subplots(figsize=(15,6))
bar6 = sns.barplot(x='country',y='deaths',data=top10_deaths,palette='bright')
plt.title('Top 10 Countries With the Highest Number of COVID-19 Deaths',fontsize=16)
plt.xlabel('Country',fontsize=12)
plt.ylabel('Number of COVID-19 Deaths',fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
for index, row in top10_deaths.iterrows():
    bar6.text(row.name,row.deaths,row.deaths,ha="center",fontsize=12)

# Display
plt.show()
```

When we look at the total number of global COVID-19 related deaths, we see that the US has the highest number of deaths.  
However, as with the local cases per province, we also need to account for the population size per country.
